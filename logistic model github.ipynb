```python
#import all libraries
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split 
import seaborn as sns
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn import datasets     #CHECKS HOW MUCH EACH VARIABLE IS IMPACTING TO OVERALAL MODEL
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression     #rfe is recursive feature extraction
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 550)
pd.set_option('display.max_colwidth', 500)
```

### import csv file

```python
#import dataset
data = pd.read_csv('Banking.txt', header=0)
data
```

### assign dependent column to new variable and drop it from original dataset using .drop()

```python
Y= data['y']
data=data.drop(['y'],axis=1)
```

### seperate all categorical data using select_dtypes(include=['object'])

```python
cat = data.select_dtypes(include=['object']).copy(deep=False)
```

```python
cat
```

```python
L=data.copy(deep=False)
L
```

### convert using one hot encoding

```python
L=pd.get_dummies(L,columns=cat.columns)
```

```python
L
```

```python
L.shape
```

### recursive feature elimination method to removee weakest features and make model more accurate

```python
logreg = LogisticRegression()
rfe = RFE(logreg, 18)  #out of 63,you want best 18 to give more accuracy
rfe = rfe.fit(L,Y )
rfe.support_
rfe.ranking_
```

### seperate best independent feature and scale it

```python
#1z are true values. it is in position 5,9,12,,16,19,,35,48,47,46,45,54,53,52,51,57,61,62,63  and correspodinf columns include "previous", "euribor3m", "job_blue-collar", "job_retired", "job_services", "job_student", "default_no", "month_aug", "month_dec", "month_jul", "month_nov", "month_oct", "month_sep", "day_of_week_fri", "day_of_week_wed", "poutcome_failure", "poutcome_nonexistent", "poutcome_success".
```

```python
col= ["previous", "euribor3m", "job_blue-collar", "job_retired", "job_services", "job_student", "default_no", "month_aug", "month_dec", "month_jul", "month_nov", "month_oct", "month_sep", "day_of_week_fri", "day_of_week_wed", "poutcome_failure", "poutcome_nonexistent", "poutcome_success"]
```

```python
X=L[col]
```

```python
Xs=scale(X)
Xs
```

### build model by splitting data into train(30%) and test(70%)

```python
X_train, X_test, Y_train, Y_test = train_test_split(Xs ,Y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)   #fit the model
```

```python
Y_pred = logreg.predict(X_test)
```

### model validation using kfold 

```python
kfold = model_selection.KFold(n_splits=10, random_state=7)
modelCV = LogisticRegression()
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, Y_train, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```

### confusion matrix to measure performance of this algorithm

```python
confusion_matrix = confusion_matrix(Y_test, Y_pred)
confusion_matrix
```

```python
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, Y_test)))
```

### classification report -           to measure quality of predictions(number of true,false)

```python
print(classification_report(Y_test, Y_pred))     
```
